## a. What was the percentage increase in the price of oil between 1998 and 2006?

## b.If the total oil supply had grown at the same rate as the former USSR oil supply between 2000 and 2006 how much would this have exceeded the demand in 2006?

```python
data
```

```python
#change_in_supply = (supply in 2006/supply in 2000)*100
change_in_supply= (data['2006'].iloc[10]/data['2000'].iloc[10])* 100
print(change_in_supply)
```

```python
#supply_in_2006 = total supply in 2000 x change in supply for USSR 2000-2006
supply_in_2006 = (data['2000'].iloc[12])*(change_in_supply/100)
print(supply_in_2006)
```

```python
#difference = supply_in_2006 - demand in 2006
difference = supply_in_2006 - data['2006'].iloc[6]
print(difference)
```

## c.In which year was the largest combined change in supply from the previous year for OPEC, the former USSR and other Non-OECD countries?

```python
data
```

```python
change_data = data.iloc[[9,10,11]]
change_data
```

```python
change_data.loc['Total'] = (change_data.sum(axis=0))
change_data
```

```python
change_data=change_data.replace(to_replace ="OPEC totalFormar USSROther non-OECD",
                 value ="-")
change_data
```

```python
change_data.loc['Combined Change'] = (change_data).sum(axis=0)
change_data['1998'].iloc[4] = "-"
change_data['1999'].iloc[4]=change_data['1999'].iloc[3]-change_data['1998'].iloc[3]
change_data['2000'].iloc[4]=change_data['2000'].iloc[3]-change_data['1999'].iloc[3]
change_data['2001'].iloc[4]=change_data['2001'].iloc[3]-change_data['2000'].iloc[3]
change_data['2002'].iloc[4]=change_data['2002'].iloc[3]-change_data['2001'].iloc[3]
change_data['2003'].iloc[4]=change_data['2003'].iloc[3]-change_data['2002'].iloc[3]
change_data['2004'].iloc[4]=change_data['2004'].iloc[3]-change_data['2003'].iloc[3]
change_data['2005'].iloc[4]=change_data['2005'].iloc[3]-change_data['2004'].iloc[3]
change_data['2006'].iloc[4]=change_data['2006'].iloc[3]-change_data['2005'].iloc[3]
change_data
```

## Since the highest combined change is 3.8 in year 2004. So the year is 2004
